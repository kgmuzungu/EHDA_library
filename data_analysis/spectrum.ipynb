```python
import pandas as pd
import matplotlib.pyplot as plot
```

```python

df = pd.read_feather('Data/corrected_database.feather')
```

```python
sample_chosen = 3000

s2 = df['current'][sample_chosen]
samplingFrequency = 100000

print(df['spray_mode'][sample_chosen])

plot.plot(s2)
plot.show()
plot.xlabel('Sample')
plot.ylabel('Amplitude')

powerSpectrum, freqenciesFound, time, imageAxis = plot.specgram(s2, Fs=samplingFrequency)
plot.xlabel('Time')
plot.ylabel('Frequency')
plot.show() 
```

```python
from scipy import signal
import numpy as np

sample_chosen = 10
M = 1024
sample = df['current'][sample_chosen]
print(df['spray_mode'][sample_chosen])
samplingFrequency = 100000

freqs, times, Sx = signal.spectrogram(sample, fs=samplingFrequency, window='hamming',
                                      nperseg=5000, scaling='spectrum')

f, ax = plot.subplots(figsize=(4.8, 2.4))
ax.pcolormesh(times, freqs / 1000, 10 * np.log10(Sx), cmap='viridis')
ax.set_ylabel('Frequency [kHz]')
ax.set_xlabel('Time [s]')
ax.set_ylim(top=1)
```

```python
sample_chosen = 500
M = 1024

samplingFrequency = 100000

for sample_chosen in df.index:

    sample = df['current'][sample_chosen]

    freqs, times, Sx = signal.spectrogram(sample, fs=samplingFrequency, window='hamming',
                                        nperseg=5000, scaling='spectrum')

    f, ax = plot.subplots(figsize=(4.8, 2.4))
    ax.pcolormesh(times, freqs / 1000, 10 * np.log10(Sx), cmap='viridis')
    ax.set_title(df['spray_mode'][sample_chosen])
    ax.set_ylabel('Frequency [kHz]')
    ax.set_xlabel('Time [s]')
    ax.set_ylim(top=1)
    plot.savefig(f"images_19-04/{sample_chosen}_{df['spray_mode'][sample_chosen]}.png")
```
